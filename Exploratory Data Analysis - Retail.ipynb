# Importing the Libraries

# EXPLORATORY DATA  ANALYSIS

```python
# finding the pairwise correlations between the columns and visualizing using heatmaps
dataset.corr()
plt.figure(figsize=(10,5))
sns.heatmap(dataset.corr(), annot = True)
plt.show()
```

```python
sns.countplot("Segment", data = dataset)
```

*Consumer contributes highest to sales whereas Home Office contributes lowest to sales.*

**Visualising the categories**

```python
# computing top categories in terms of sales
top_category_s = dataset.groupby('Category').Sales.sum().nlargest(keep = "first")
# computing top categories in terms of profit
top_category_p = dataset.groupby('Category').Profit.sum().nlargest(keep = "first")

#plotting to see it visually
plt.style.use('seaborn')
top_category_s.plot(kind = 'bar', figsize = (10,5), fontsize = 14)
top_category_p.plot(kind = 'bar', figsize = (10,5), fontsize = 14, color = 'red')
plt.xlabel('Category', fontsize = 15)
plt.ylabel('Total Sales/Profit', fontsize = 15)
plt.title('Top Category Sales vs Profit', fontsize = 15)
plt.show()
```

**Visualising the Sub-categories**

```python
# computing top sub-categories in terms of sales
top_subcategory_s = dataset.groupby('Sub-Category').Sales.sum().nlargest(n = 9994, keep = "first")
# computing top sub-categories in terms of profit
top_subcategory_p = dataset.groupby('Sub-Category').Profit.sum().nlargest(n = 9994, keep = "first")

# plotting to see it visually
plt.style.use('seaborn')
top_subcategory_s.plot(kind = 'bar', figsize = (10,15), fontsize = 14)
top_subcategory_p.plot(kind = 'bar', figsize = (10,15), fontsize = 14, color = 'red')
plt.xlabel('Sub-Category', fontsize = 15)
plt.ylabel('Total Sales/Profit', fontsize = 15)
plt.title('Top Sub-Category Sales vs Profit', fontsize = 15)
plt.show()
```

```python
# A more detailed view
plt.figure(figsize = (14,12))
statewise = dataset.groupby(['Sub-Category'])['Profit'].sum().nlargest(n = 9994, keep = "first")
statewise.plot.barh(fontsize = 14) # h for horizontal
```

*The above graph clearly shows that Copiers and Phones have the highest sales and profit whereas Tables have neagtive profit*

**Visualising the Discount**

```python
plt.figure(figsize = (8,7))
sns.lineplot(dataset['Discount'], dataset['Profit'], data = dataset)
```

**Visualizing Sales vs Profit in different Regions**

```python
plt.figure(figsize = (6,6))
plt.title('Region', fontsize = 15)
plt.pie(dataset['Region'].value_counts(), labels = dataset['Region'].value_counts().index, autopct = '%1.1f%%', textprops = {'fontsize':15})
plt.show()
```

*The pie chart shows that West region is the highest contributor while South region is the lowest contributor to sales*

```python
# computing states in terms of sales
top_states_s = dataset.groupby('State').Sales.sum().nlargest(n = 50, keep = "first")
# computing states in terms of profit
top_states_p = dataset.groupby('State').Profit.sum().nlargest(n = 50, keep = "first")

# plotting to see it visually
plt.style.use('seaborn')
top_states_s.plot(kind = 'bar', figsize = (15,15), fontsize = 12)
top_states_p.plot(kind = 'bar', figsize = (15,15), fontsize = 12, color = 'red')
plt.xlabel('States', fontsize = 15)
plt.ylabel('Total Sales', fontsize = 15)
plt.title('50 States Sales vs Profit', fontsize = 15)
plt.show()
```

```python
plt.style.use('seaborn')
dataset.plot(kind = 'scatter', figsize = (10,5), x = 'Sales', y = 'Profit', c = 'Discount', s = 20, fontsize = 16, colormap = 'viridis')
plt.ylabel('Total Profits', fontsize = 16)
plt.title('Interdependency of Sales, Profits and Discounts', fontsize = 16)
plt.show()
```

*The graph clearly shows that if we give more discount on products, sales increases but profit decreases.*
